## filtering the place ids
The metro state year over year contains a single place id field (Place Id with less than 3 digits are state level data). This maps to the census delination data in either one of two fields, the census delination data CBSA Code or Metropolitan Division Code.

We need get the pair of FIPS State Code,FIPS County Code from the delination file where the metro data Place Id matches either the census CBSA Code OR Metropolitan Division Code. 

1. get the unique Place Ids from the metro data with values > 100
2. filter the census delination data where the CBSA Code or the Metro Division Code matches one of the unique metro place ids
3. get the unique covid_fips (an augmented column in the census data) codes

```python
metro_cbsa_ids = metro_housing_df[metro_housing_df["Place ID"] >= 100]["Place ID"].unique()
metro_delination_df = census_delination_df[(census_delination_df["CBSA Code"].isin(metro_cbsa_ids)) | (census_delination_df["Metropolitan Division Code"].isin(metro_cbsa_ids))]
census_fips_codes = metro_delination_df["covid_fips"].unique()
```

```python
metro_covid_data = raw_covid_df[raw_covid_df["fips"].isin(census_fips_codes)]
```

```python
metro_covid_data
```

```python
# now get the data at the end of each quarter 
quarter_dates = np.array(["2020-03-31","2020-06-30","2020-09-30","2020-12-31"])
metro_quarter_data_df = metro_covid_data[metro_covid_data["date"].isin(quarter_dates)]
```

```python
metro_quarter_data_df
```

```python
def date_to_quarter(x):
    """
    Convert a date string into the quarter number
    """
    result = np.where(quarter_dates == x)
    return result[0][0] +1 
metro_quarter_data_df["quarter"] = metro_quarter_data_df["date"].apply(date_to_quarter)
```

```python
covid_metro_quarter_data_df = metro_quarter_data_df.merge(metro_delination_df, left_on="fips", right_on="covid_fips")
```

```python
covid_metro_quarter_data_df
```

```python
metro_housing_df
```

## add in population data
for each fips value, add in the population data

```python
population_df = pd.read_csv("../covid_data/data/census_population_only_estimates_2019.csv")
```

```python
population_df["covid_fips"] = population_df["STATE"]*1000 + population_df["COUNTY"]
```

```python
population_df
```

```python
covid_metro_quarter_with_placeids_population_df = covid_metro_quarter_with_placeids_df.merge(population_df, left_on="fips", right_on="covid_fips")
```

```python
covid_metro_quarter_with_placeids_population_df
```

```python
covid_metro_quarter_with_placeids_population_agg_by_date_quarter_placeid_df = covid_metro_quarter_with_placeids_population_df.groupby(["date","quarter","census_place_id"], as_index=False).agg({"cases": {"sum"}, "POPESTIMATE2019": {"sum"}})
```

## final covid data
We now have a data frame with the total cases at the end of each quarter, and the population estimate for each place_id from the housing data

```python
covid_metro_quarter_with_placeids_population_agg_by_date_quarter_placeid_df
```

## add covid data to the metro housing data
now we want to merge in this covid aggregate data into the housing data based place id and quarter

```python
metro_housing_with_covid = metro_housing_df.merge(covid_metro_quarter_with_placeids_population_agg_by_date_quarter_placeid_df, left_on=["Place ID","Quarter"], right_on=["census_place_id","quarter"])
```

```python
metro_housing_with_covid.info(verbose=True)
```

```python
metro_housing_with_covid = metro_housing_with_covid.rename(columns={("cases", "sum"): "total_covid_cases", ("POPESTIMATE2019","sum"): "population"})
```

```python
metro_housing_with_covid.info(verbose=True)
```

## Next
Now that we have the total covid cases number at the end of each quarter assigned to each place id in the metro housing data, we want to calculate the new cases from quarter to quarter for each place id. quarter 1 the new cases, should just be the total cases for that quarter. for quarter 2, the new cases should be q2 total cases - q1 total cases, and so forth.

```python
metro_housing_with_covid.groupby(by=["Place ID"]).head(10)
```

```python
# https://stackoverflow.com/questions/20648346/computing-diffs-within-groups-of-a-dataframe
metro_housing_with_covid = metro_housing_with_covid.sort_values(by=["Place ID","Quarter"])
```

```python
metro_housing_with_covid['new_cases'] = metro_housing_with_covid.groupby("Place ID")["total_covid_cases"].diff()
```

```python
metro_housing_with_covid
```

```python
metro_housing_with_covid["new_cases"] = metro_housing_with_covid["new_cases"].fillna(metro_housing_with_covid["total_covid_cases"])
```

```python
metro_housing_with_covid
```

## now scale by population
we want to scale the cases numbers by a measure of the population

```python
metro_housing_with_covid["total_covid_per_capita_1000"] = metro_housing_with_covid["total_covid_cases"]/(metro_housing_with_covid["population"]/1000)
```

```python
metro_housing_with_covid["new_cases_per_capita_1000"] = metro_housing_with_covid["new_cases"]/(metro_housing_with_covid["population"]/1000)
```
