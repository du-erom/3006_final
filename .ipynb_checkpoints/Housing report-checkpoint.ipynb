# Introduction

The COVID-19 pandemic has altered much in the daily lives of millions of people.  News reports have reported people leaving high density metro areas for suburbs and rural areas.  This project seeks to identify changes in US housing market data since March 2020 and investigates correlations with covid prevalence.

This project relies on data from the New York Times' COVID data set and the Federal Housing Finance Agency (FHFA) as our reliable data sources.



# Market Level Changes in the HPI in 2020

The FHFA reports a Housing Price Index (HPI) for every state and most metropolitan areas in the US every quarter.  Additionally, they report an HPI for the non-metro area for each state. 

To begin our analysis of the housing market, we need a good measure of change in the index for each reported area over the time since the start of the COVID pandemic.  Since our housing data is quarterly, and policy interventions began in mid-March 2020 we will look for changes in the housing market from Q2 2020 to the most recent data in the data set.

The FHFA corrects some of its data for seasonality, suggesting that there are some seasonal effects that could confuse our analysis. Because the data we are using has not been corrected, we will compare changes in the HPI as a year over year same quarter data comparison.

To begin, we will load the housing module. We will use that module to create a housing data object with our subset of data since 2015.  We will use our data object and methods from our module create arrays with the HPI, and year over year change in both change in the index and the percent change of the index. 

```python
import housing
import numpy as np
import matplotlib.pyplot as plt

o = housing.HousingData() #create a data object with HPI data between 2015 and 2020
state_change = housing.yoy_change(o.state_data, 2020) #calculate yoy change in 2020 at state level
msa_change = housing.yoy_change(o.metro_data, 2020) #calculate yoy change in 2020 at metro level
```

Now that we have data for 2020 let's look at this from a market level,  with a histogram of the year over year change in the HPI for state level data. 

```python
d_ch = [] #instantiate an empty list
for i in state_change:
    d_ch.append(i[5]) #append yoy change in HPI to list

d = d_ch #rename the list
plt.figure(figsize =(12,8))
plt.hist(d, bins =40) #create a histogram with 40 bins
plt.title("Histogram of Year Over Year Change in 2020 in HPI at a State level") #title the plot
plt.xlabel('Change in HPI') #label the x-axis

plt.show()
```

```python
p_ch = [] #instatiate empty list
for i in state_change: 
    p_ch.append(i[6]) #append percent year over year change to list

p = p_ch #change name of list
plt.figure(figsize =(12,8))
plt.hist(p, bins =40)#create a histogram with 40 bins
plt.title("Histogram of Year Over Year Percentage Change in 2020 in HPI at a State level")
plt.xlabel('Percent Change in HPI')

plt.show()
```

It looks like we have year over year growth for all state in all quarters of 2020, and that growth is distributed in a log normal fashion.  Let's look at the percentage growth for the same period.

When we look at percentages we get closer to a normal distribution, though it still looks skewed.

Let's look at how the metro level changes in HPI are distributed.



```python
md_ch=[] #instantiate list
for i in msa_change:
    md_ch.append(i[5])#append change in HPI

md = md_ch
plt.figure(figsize =(12,8))
plt.hist(md, bins =100) #histogram with 100 bins
plt.title("Histogram of Year Over Year Change in HPI in 2020 at a metro level")
plt.xlabel('Change in HPI') #label the x-axis

plt.show()
```

This distribution is also skewed, but we can see some areas of negative changes in this distribution.  

Let's see how significant that negative change is on a percentage basis.

```python
mp_ch = []

for i in msa_change:
    mp_ch.append(i[6])

mp = mp_ch
plt.figure(figsize =(12,8))
plt.hist(mp, bins =80)

plt.title("Histogram of Year Over Year Percentage Change in HPI in 2020 at a metro level")
plt.xlabel('Percent Change in HPI')

plt.show()
```

This distribution looks very close to a normal distribution.  The negative growth we saw in the point change of HPI is present here,  the largest drop in HPI is around 7% year over year.

These  histograms tell us that some metro areas experienced a drop in home prices in 2020 while on a state level there was only growth in home prices, but tell us nothing of the context of these changes.  To add this context we will prepare our data to show the average changes on both a state and metro levels quarter by quarter since 2016 in a series of line plots.  We will also prepare ridgeline plots of these histograms for a greater historical context.  We will look at five years of data and build a ridgeline style histogram for each quarter on a single plot. 

To do this we have to first pull the appropriate data from our data object. We will pull year over year data from 2016 to 2019 for state and metro level data, and append that data to our current arrays.

```python
import pandas as pd
#calculate year over year changes at state level from 2016 to 2020
s_2016 = housing.yoy_change(o.state_data, 2016)
s_2017 = housing.yoy_change(o.state_data, 2017)
s_2018 = housing.yoy_change(o.state_data, 2018)
s_2019 = housing.yoy_change(o.state_data, 2019)
s_2020 = housing.yoy_change(o.state_data, 2020)
state = []#instantiate an empty list for year over year data at state level 
#add year over year data to state list
for line in s_2016:
    state.append(line)
for line in s_2017:
    state.append(line)
for line in s_2018:
    state.append(line)
for line in s_2019:
    state.append(line)
for line in s_2020:
    state.append(line)
#calculate year over year changes at metro level from 2016 to 2020    
m_2016 = housing.yoy_change(o.metro_data, 2016)
m_2017 = housing.yoy_change(o.metro_data, 2017)
m_2018 = housing.yoy_change(o.metro_data, 2018)
m_2019 = housing.yoy_change(o.metro_data, 2019)
m_2020 = housing.yoy_change(o.metro_data, 2020)
msa = []#instantiate an empty list for year over year data at metro level
#add year over year data to msa list
for line in m_2016:
    msa.append(line)
for line in m_2017:
    msa.append(line)
for line in m_2018:
    msa.append(line)
for line in m_2019:
    msa.append(line)
for line in m_2020:
    msa.append(line)
#turn state list to pandas dataframe
df_s = pd.DataFrame(state, 
                    columns = ['Place Name', 'Place ID', 'Year', 'Quarter',\
                              'HPI', 'YoY change', '% YoY change', 'year_dec_quarter'])
#turn metro list to dataframe
df_m = pd.DataFrame(msa, 
                    columns = ['Place Name', 'Place ID', 'Year', 'Quarter',\
                               'HPI', 'YoY change', '% YoY change', 'year_dec_quarter'])

```

# Recent Historic Context for Market Level Changes in HPI

Now that we have year over year changes in HPI data since 2016 in a data structure, we can leverage NumPy's aggregation to find the average change by quarter and create plots to put the changes in 2020 in context.

```python
#state averages by quarter
#numpy arrays for the percent change and quarter year
ps_change = df_s[['% YoY change']].values
qs_year = df_s[['year_dec_quarter']].values
#list of unique quarters
all_qs = df_s['year_dec_quarter'].unique()
#list of tuples with decimal quarter, average % year over year state level growth
s_avgs = [(i, ps_change[qs_year == i].mean()) for i in all_qs]

#metro averages by quarter
#numpy arrays for the percent change and quarter year
pm_change = df_m[['% YoY change']].values
qm_year = df_m[['year_dec_quarter']].values
#list ofunique quarters
all_qm = df_m['year_dec_quarter'].unique()
#list of tuples with decimal quarter, average % year over year metro level growth
m_avgs = [(i, pm_change[qm_year == i].mean()) for i in all_qs]


#create lists for x and y
x = []
y = []
#iterate through the dictionary so x = year, y = % year over year growth
for tu in s_avgs:
    x.append(tu[0])
    y.append(tu[1])
#create a line plot
plt.figure(figsize =(12,8))
plt.plot(x,y, color ='b')
plt.xlabel('Year')
plt.ylabel('Pecent Change')
plt.title('Year Over Year Mean Percenage Change in HPI in State Data')
plt.show()
```

The plot above shows that the steepest decline in state level HPI growth on a percentage basis happened in the second quarter of 2020.  We see a dramatic rebound in the growth in home prices over the third and fourth quarter of the year.  As the incident rate of COVID was highest in the fourth quarter of 2020, but the policy response was most restrictive in the second quarter of 2020, the change in HPI at the state level might be correlated more to policy response than an individual response to the novel coronavirus.

Now lets look at the same plot for the HPI on a metro level.

```python
#create lists for x and y
x = []
y = []
#iterate through the dictionary so x = year, y = % year over year growth
for tu in m_avgs:
    x.append(tu[0])
    y.append(tu[1])
#create a scatter plot
plt.figure(figsize =(12,8))
plt.plot(x,y, color ='b')
plt.xlabel('Year')
plt.ylabel('Pecent Change')
plt.title('Year Over Year Mean Percenage Change in HPI in Metro Data')
plt.show()
```

The above plot shows the steepest drop in the percent of HPI aggregated at the metropolitan level in Q2 2020.  But show the steepest increase in HPI on a percentage basis in Q4 2020. Which is exactly what we saw in the state level data.  

The percent changes in average HPI tell us how the US housing market as a whole reacted to the COVID-19 pandemic *on average*.  Aggregating  to averages can lead us to miss many of the nuanced changes in the data.  While the volume of data means it is difficult  to visualize all the data on a plot, we can get more information about market changes by comparing quarter by quarter densities (similar to histograms) together in ridgeline  plots.  To prepare for these plots we need to import a library that leverages matplotlib called joypy.  

```python
#import libraries
import joypy 
from matplotlib import cm
#show plot without calling show
%matplotlib inline 

#only label the year, not the quarter
labs = [2016, None, None, None, 2017, None, None, None, 2018, None, None, None, 2019, None, None, None, 2020, None, None, None]

#create ridgeline plot of state data for YoY change in HPI
fig, axes =joypy.joyplot(df_s, by = 'year_dec_quarter', 
       column = 'YoY change', labels = labs, range_style = 'own', ylim='own',
       figsize = (12,8), grid = True, colormap = cm.viridis, 
       title = 'YoY Change in HPI Agregated by State', alpha = 0.5)
plt.xlabel('YoY Change in HPI')
```

The above plot shows the distribution of change in year over year HPI on a state level through the end of 2020.  We can see a shift to the left in the distribution going from Q1 2020 to Q2 2020.  As Q2 2020 was the first full quarter with the pandemic, we can confidently say that the slowing in growth in that quarter was due to the pandemic, though we cannot pinpoint the precise proximate cause.

Let's look at this plot in terms of year over year percentage change in HPI.

```python

#create ridgeline plot of state data for YoY Percentage change in HPI
fig, axes =joypy.joyplot(df_s, by = 'year_dec_quarter', 
       column = '% YoY change', labels = labs, range_style = 'own', ylim='own',
       figsize = (12,8), grid = True, colormap = cm.viridis, 
       title = 'Percentage YoY Change in HPI Agrigated by State', alpha = 0.5)

plt.xlabel('YoY % Change in HPI')
```

In the above plot the Q4 2020 grabs our attention first, as it is front and center and looks like a classic Gaussian distribution.   Closer investigation shows a skew in the upper tail of the distribution.  

Looking at Q2 2020, we can see the changes in the market did not just shifting the distribution to the left.  They also reduced the standard variation.  

In Q3 2020 we see that the mean shifted back to the right, but the lower bound of the density moved left.  This tells us that the growth was unevenly distributed enough that even at the state level of aggregation we can see some areas with very low or even negative growth.  

Now let's look at the metro level data in these ridgeline plots.

```python
#create ridgeline plot of metro data for YoY change in HPI
fig, axes =joypy.joyplot(df_m, by = 'year_dec_quarter', 
       column = 'YoY change', labels = labs, range_style = 'own', ylim='own',
       figsize = (12,8), grid = True, colormap = cm.viridis, 
       title = 'YoY Change in HPI Agregated by Metro Area', alpha = 0.5)
plt.xlabel('YoY Change in HPI')
```

Here, the feature of Q4 2020 that stands out is how wide the tails of the distribution are; the lower bound of the change in HPI is the lowest seen in the data set.  

In Q2 2020 we can see the average and the lower bound of the distribution moved left, while the upper bound did not change significantly.  As this feature is not present in the state level data this might suggests that there are local variations within states.  While these variations might be correlated to the variation in infection rates, it also could be conflated to variations in the policy responses by cities.  Because this plot is showing the raw change in HPI, this could just be artifact of some areas with especially high HPI value (like New York City and San Francisco) experienced some of the most drastic policy changes.

In Q3 we see distribution move to the right and the standard variation go down because the distribution narrows.

Let's see how the plot changes when we plot the percent change in HPI.

```python
#create ridgeline plot of metro data for percentage YoY change in HPI
fig, axes =joypy.joyplot(df_m, by = 'year_dec_quarter', 
       column = '% YoY change',labels = labs, range_style = 'own', ylim='own',
       figsize = (12,8), grid = True, colormap = cm.viridis, 
       title = 'Percentage YoY Change in HPI Agrigated by Metro Area', alpha = 0.5)

plt.xlabel('YoY % Change in HPI')
```

Here, we see evidence that the hypothesis we put forward to explain the lower bound of Q4 2020 in the change of HPI plot might be correct.  On a percentage basis the lower bound of the Q4 2020 distribution is not as low as we saw in Q1 2019.  This suggests that the negative change in HPI for Q4 2020 was in areas with a high baseline HPI.
